```python
import requests
from bs4 import BeautifulSoup

def GetWebScrape(url, segment, class_tag, element_tags):
  if not url or not segment or not class_tag or not element_tags:
    return

  response = requests.get(url)

  if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    segment = soup.find('div', {'id': segment})

    if segment:
      class_elements = segment.find_all(class_tag)
      extracted_class = [element.get_text(strip=True) for element in class_elements
                                                        if element.get_text(strip=True)][0]
      text_elements = segment.find_all(element_tags)
      extracted_texts = [element.get_text(strip=True) for element in text_elements
                                                        if element.get_text(strip=True)]

      return extracted_class, extracted_texts
    else:
      return
  else:
    return
```

```python
def remove_unnecessary_columns(life_styles_extracteds, column_number):
  life_styles_extracteds_updated = {}

  for key, value in life_styles_extracteds:
    if value is not None: #and isinstance(value, (list, tuple)) and len(value) > 1:
      if isinstance(value[1], list) and len(value[1]) >= column_number:
        print("if")
        life_styles_extracteds_updated[key] = value[0], value[1][column_number:]
      elif isinstance(value[1], list): # If it's a list but has less than column_number elements, keep the list as is
        print("elif")
        life_styles_extracteds_updated[key] = value[0], value[1]
      else: # If the second element is not a list, keep the original value
        print("else")
        life_styles_extracteds_updated[key] = value
      #else:
      #  life_styles_extracteds_updated[key] = value[0], value[1]
    else:
      life_styles_extracteds_updated[key] = None

  return life_styles_extracteds_updated
```

```python
def concatenate_and_replace(key, data_list):
  if not data_list or len(data_list) <= 5:
    return data_list

  if key not in excluded_keys:
    concatenated_string = f"{data_list[0]}-{data_list[1]}-{data_list[2]}-{data_list[3]}-{data_list[4]}"
    concatenated_list = [concatenated_string.replace("  ", "").replace("\n","")]
    concatenated_list.extend(data_list[5:])
    new_list = concatenated_list
  elif key in excluded_keys:
    concatenated_string = f"{data_list[1]}-{data_list[2]}-{data_list[3]}-{data_list[4]}-{data_list[5]}"
    concatenated_list = [concatenated_string.replace("  ", "").replace("\n","")]
    concatenated_list.extend(data_list[6:])
    concatenated_list.append(data_list[0])
    new_list = concatenated_list

  return new_list
```

```python
from transformers import AutoTokenizer
import torch
from FlagEmbedding import BGEM3FlagModel

class BGEM3SentenceEmbedder:
    def __init__(
        self,
        model_name: str = "BAAI/bge-m3",
        device: str     = None,
        max_length: int = 8192
    ):
        self.device     = device or ("cuda" if torch.cuda.is_available() else "cpu")
        self.tokenizer  = AutoTokenizer.from_pretrained(model_name, use_fast=True)
        self.model      = BGEM3FlagModel(model_name) # BGEM3FlagModel.from_pretrained(model_name).to(self.device)
        #self.model.eval()
        self.max_length = max_length

    @torch.no_grad()
    def encode(self, texts, batch_size: int = 32, return_sparse: bool = True):
        """
        texts: list[str] or str
        returns: torch.Tensor or scipy.sparse matrix (depending on return_sparse)
        """
        if isinstance(texts, str):
            texts = [texts]

        out = self.model.encode(
            sentences=texts,
            batch_size=batch_size,
            return_dense=True,
            return_sparse=return_sparse
        )
        return out
```

```python
import umap.umap_ as umap


def convert_to_2d(dense_vecs, n_neighbors=15):
  reducer = umap.UMAP(n_neighbors=n_neighbors,
                      n_components=2,
                      metric='cosine',
                      min_dist=0.1,
                      random_state=42)
  embeddings_2d = reducer.fit_transform(dense_vecs)
  return embeddings_2d
```

```python
import matplotlib.pyplot as plt
from adjustText import adjust_text
from sklearn.cluster import KMeans


# in this step, we find the optimal k value
inertia = []
k_range = range(1, 15)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    embeddings_2d = convert_to_2d(dense_vecs)
    kmeans.fit(embeddings_2d)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


optimal_k=10
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
embeddings_2d = convert_to_2d(dense_vecs)
labels = kmeans.fit_predict(embeddings_2d)

plt.figure(figsize=(10, 8))
#scatter = plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=labels, cmap='Spectral', s=50)
scatter = plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], c=labels, cmap='tab10', s=80, alpha=0.7, edgecolor='k')

# Using adjustText for improved text visibility
texts = []
for i, name in enumerate(life_styles_names):
    texts.append(plt.text(embeddings_2d[i, 0], embeddings_2d[i, 1], name, fontsize=8))

adjust_text(texts, arrowprops=dict(arrowstyle='-', color='gray', lw=0.5))

# Colorbar with clear cluster labels
cbar = plt.colorbar(scatter, ticks=range(optimal_k))
cbar.set_label('Clusters', fontsize=14)
cbar.ax.set_yticklabels([f'Cluster {i}' for i in range(optimal_k)], fontsize=12)

#plt.colorbar(label='Clusters')
plt.title('UMAP + KMeans Clustering')
plt.xlabel('Dim-1')
plt.ylabel('Dim-2')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

# Similarity

```python
import numpy as np

def cosine_similarity(a, b):
    a = np.asarray(a)
    b = np.asarray(b)

    if a.ndim > 1:
        a = a[0]

    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    if norm_a == 0 or norm_b == 0:
        return 0.0
    return np.dot(a, b) / (norm_a * norm_b)
```

```python
import numpy as np

try:
    from FlagEmbedding import BGEM3FlagModel
except ImportError:
    !pip install FlagEmbedding==1.3.3
    from FlagEmbedding import BGEM3FlagModel

life_styles_embeddings=life_styles_embeddings

def get_similar_lifestyles(tweet, k=3):
    if not tweet or life_styles_embeddings is None:
        return []

    tweet_output = model.encode([tweet], return_dense=True, return_sparse=True)
    tweet_dense = tweet_output["dense_vecs"][0]
    tweet_sparse = tweet_output["lexical_weights"][0]

    dense_similarities = []
    sparse_similarities = []
    life_style_names_list = []

    for name, embedding_result in life_styles_embeddings.items():
      if embedding_result is not None and "dense_vecs" in embedding_result and "lexical_weights" in embedding_result:
        life_style_dense = embedding_result["dense_vecs"][0]
        life_style_sparse = embedding_result["lexical_weights"][0]

        dense_similarities.append(cosine_similarity(tweet_dense, life_style_dense))
        sparse_similarities.append(model.compute_lexical_matching_score(tweet_sparse, life_style_sparse))
        life_style_names_list.append(name)

    if not dense_similarities or not sparse_similarities:
      print("No valid lifestyle embeddings with both dense and sparse vectors found for deep similarity calculation.")
      return []

    combined_similarities = [
        d * 0.75 + s * 0.25 for d, s in zip(dense_similarities, sparse_similarities)
    ]

    k = min(k, len(combined_similarities))

    most_similar_indices = np.argsort(combined_similarities)[-k:]
    most_similar_indices = most_similar_indices[::-1]

    most_similar_characters = [life_style_names_list[i] for i in most_similar_indices]

    return f"The tweet is most similar to (top {k}): {', '.join(most_similar_characters)}"
```

```python
life_styles_embeddings=life_styles_embeddings


def get_closest_lifestyle(survey_answer):
  if not survey_answer or life_styles_embeddings is None:
    print("Survey answers or lifestyle embeddings are missing.")
    return None


  survey_answer = [survey_answer]

  try:
    survey_outputs = model.encode(survey_answer, return_dense=True, return_sparse=True)
  except Exception as e:
    print(f"Error encoding survey answers: {e}")
    return None


  survey_dense_embeddings = survey_outputs.get("dense_vecs")
  if survey_dense_embeddings is None or len(survey_dense_embeddings) == 0:
      print("Could not get dense embeddings for survey answers.")
      return None

  tweet_dense = survey_dense_embeddings[0]

  survey_sparse_embeddings = survey_outputs.get("lexical_weights")
  if survey_sparse_embeddings is None or len(survey_sparse_embeddings) == 0:
      print("Could not get sparse embeddings for survey answers.")
      return None


  # Calculate the combined sparse embedding by averaging the lexical weights
  combined_survey_sparse_embedding = {}
  if isinstance(survey_sparse_embeddings, list):
    valid_sparse_embeddings = [d for d in survey_sparse_embeddings if isinstance(d, dict)]
    if not valid_sparse_embeddings:
        print("No valid sparse embeddings found in survey_outputs.")
        return None
    all_keys = set().union(*(d.keys() for d in valid_sparse_embeddings))
    for key in all_keys:
      combined_survey_sparse_embedding[key] = np.mean([d.get(key, 0) for d in valid_sparse_embeddings])
  elif isinstance(survey_sparse_embeddings, dict):
    combined_survey_sparse_embedding = survey_sparse_embeddings
  else:
      print("Survey sparse embeddings are in an unexpected format.")
      return None


  all_life_styles_dense_vecs = []
  all_life_styles_sparse_vecs = []
  life_style_names_list = []

  for name, embedding_result in life_styles_embeddings.items():
    if embedding_result is not None and isinstance(embedding_result, dict) and "dense_vecs" in embedding_result and "lexical_weights" in embedding_result and embedding_result["dense_vecs"] is not None and embedding_result["lexical_weights"] is not None:
      if len(embedding_result["dense_vecs"]) > 0 and len(embedding_result["lexical_weights"]) > 0:
        life_style_dense_vec = embedding_result["dense_vecs"][0]
        life_style_sparse = embedding_result["lexical_weights"][0]

        all_life_styles_dense_vecs.append(life_style_dense_vec)
        all_life_styles_sparse_vecs.append(life_style_sparse)
        life_style_names_list.append(name)
      #else:
       #   print(f"Skipping '{name}' due to empty dense_vecs or lexical_weights lists.")
    #else:
      #  print(f"Skipping '{name}' due to missing or invalid embedding structure.")

  if not all_life_styles_dense_vecs or not all_life_styles_sparse_vecs or len(all_life_styles_dense_vecs) != len(all_life_styles_sparse_vecs) or len(all_life_styles_dense_vecs) != len(life_style_names_list):
    print("No valid lifestyle embeddings found for similarity calculation, or mismatch in collected lists.")
    return None

  dense_similarities = [
      cosine_similarity(tweet_dense, life_dense) for life_dense in all_life_styles_dense_vecs
  ]

  sparse_similarities = [
      model.compute_lexical_matching_score(combined_survey_sparse_embedding, life_sparse) for life_sparse in all_life_styles_sparse_vecs
  ]

  if not dense_similarities or not sparse_similarities or len(dense_similarities) != len(sparse_similarities):
      print("Similarity calculation resulted in empty or mismatched lists.")
      return None

  combined_similarities = [
      d * 0.7 + s * 0.3 for d, s in zip(dense_similarities, sparse_similarities)
  ]

  if not combined_similarities:
      print("Combined similarities list is empty.")
      return None


  most_similar_index = np.argmax(combined_similarities)
  most_similar_lifestyle = life_style_names_list[most_similar_index]

  print(f"All survey answers combined are most similar to: {most_similar_lifestyle}")
  #return most_similar_lifestyle
```

```python
import tweepy
import time
import pandas as pd
from ratelimit import limits, sleep_and_retry


# Rate limit decorator: 50 calls per 15 minutes (900 seconds)
CALLS = 50
PERIOD = 900

@sleep_and_retry
@limits(calls=CALLS, period=PERIOD)
# Modify the function signature to accept arbitrary keyword arguments
def get_tweets_with_limit(user_id, max_results=50, **kwargs):
    # Pass the received keyword arguments (including pagination_token) to the client call
    return client.get_users_tweets(id=user_id,
                                   max_results=max_results,
                                   tweet_fields=["entities"],
                                   exclude=["retweets","replies"], **kwargs)
def get_user_id(username):
    try:
        user = client.get_user(username=username)
        return user.data.id
    except tweepy.TweepyException as e:
        print(f"Error getting user ID: {e}")

def get_user_tweets(username, tweet_count=20):
    if not username:
        print("Please provide a username.")
        return
    try:
      user_id = get_user_id(username)

      #tweet_fields = ["created_at", "public_metrics", "lang", "source"]

      all_tweets = []

        # The Paginator will now correctly pass the pagination_token to the decorated function
      for tweet in tweepy.Paginator(get_tweets_with_limit,
                                    user_id=user_id,
                                    max_results=tweet_count).flatten(limit=tweet_count):

          has_urls = tweet.entities and "urls" in tweet.entities
          has_media = tweet.entities and "media" in tweet.entities
          if not (has_urls or has_media):
              all_tweets.append(tweet.text)
              print(f"{tweet.text}\n")
    except tweepy.TooManyRequests:
        print("Rate limit hit. Waiting 15 minutes...")
        time.sleep(900)  # Wait 15 minutes
        # Retry the request after waiting
        for tweet in tweepy.Paginator(get_tweets_with_limit,
                                      user_id=user_id,
                                      max_results=tweet_count).flatten(limit=tweet_count):
            has_urls = tweet.entities and "urls" in tweet.entities
            has_media = tweet.entities and "media" in tweet.entities
            if not (has_urls or has_media):
                all_tweets.append(tweet.text)
                print(f"{tweet.text}\n")

    # Save to CSV
    df = pd.DataFrame(all_tweets, columns=["Text"])
    df.to_csv("text_tweets.csv", index=False)
    print("Text tweets saved to text_tweets.csv")

    return all_tweets
```
